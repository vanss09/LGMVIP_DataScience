```python
import cv2
from PIL import Image
from IPython.display import display
```

### Loading the Image

```python
img = cv2.imread("puppies-cover.jpg")
rgb_img =  cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(rgb_img))
```

### Converting Image to Greyscale Image

```python
greyscale_img = cv2.cvtColor(rgb_img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale_img))
```

### Converting the Image to Negative Image

```python
invert_img = 255 - greyscale_img
display(Image.fromarray(invert_img))
```

### Smoothening the Image

```python
blur_img = cv2.GaussianBlur(invert_img, (21,21), 0) #Blur the Image by gaussian function
display(Image.fromarray(blur_img))
```

### Invert the blurred image

```python
invert_blur_img = 255 - blur_img
display(Image.fromarray(invert_blur_img))
```

### Converting Image into Pencil Sketch 

```python
pencil_sketch_img = cv2.divide(greyscale_img, invert_blur_img, scale=256)
display(Image.fromarray(pencil_sketch_img))
```
